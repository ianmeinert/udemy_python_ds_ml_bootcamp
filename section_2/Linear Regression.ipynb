```python
import matplotlib.pyplot as plot
import pandas as pd
from sklearn.linear_model import LinearRegression
from pandas import DataFrame as df
```

```python
data = pd.read_csv('cost_revenue_clean.csv')
```

```python
data.describe()
```

```python
X = df(data, columns=['production_budget_usd'])
y = df(data, columns=['worldwide_gross_usd'])
```

```python
plot.figure(figsize=(10,6))                # chart size
plot.scatter(X, y, alpha=0.3)              # alpha is transparency of point
plot.title('Film Cost vs Global Revenue')  # chart title
plot.xlabel('Production Budget $')         # y-axis title
plot.ylabel('Worldwide Gross $')           # x=axis title
plot.ylim(0, 3000000000)                   # y-axis range limits
plot.xlim(0, 450000000)                    # x-axis range limits
plot.show()
```

```python
regression = LinearRegression()
regression.fit(X, y)
```

```python
regression.coef_                           # theta_1
```

```python
regression.intercept_                      # intercept
```

```python
plot.figure(figsize=(10,6))                                            # chart size
plot.scatter(X, y, alpha=0.3)                                          # alpha is transparency of point
plot.plot(X, regression.predict(X), color='#eeeccc', linewidth = 2)    # plot line
plot.title('Film Cost vs Global Revenue')                              # chart title
plot.xlabel('Production Budget $')                                     # y-axis title
plot.ylabel('Worldwide Gross $')                                       # x=axis title
plot.ylim(0, 3000000000)                                               # y-axis range limits
plot.xlim(0, 450000000)                                                # x-axis range limits
plot.show()
```

```python
regression.score(X, y)
```
